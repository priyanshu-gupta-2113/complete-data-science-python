## Importing Modules in Python: Modules and packages 

In python,modules and packages help organize and reuse code.

```python
import math
math.sqrt(16)
```

```python
from math import pi,sqrt
print(pi*2)
print(sqrt(25))
```

```python
## to import everything from a library 
## able to access all the functions
from math import *

print(pi)
print(e)
```

```python
import package.maths
package.maths.addition(4,5)
```

```python
from package.maths import addition
addition(3,3)
```

```python
from package import maths
maths.addition(3,5)
```

```python
from package.subpackage.hehe import multiply
multiply(3,4)
```

```python
import csv
with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['krish',32])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
from datetime import datetime,timedelta
now=datetime.now()
print(datetime.now())
yesterday=now-timedelta(days=1)
print(yesterday)
```

```python
## Regular expression - to match pattern 

import re 
pattern=r'\d+' ## d is digits
text='there are 123 apples'
match=re.search(pattern,text)
print(match.group())
```
