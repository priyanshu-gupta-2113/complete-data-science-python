```python
##Creating Tuples

##empty
t1=()
print(t1)
print(type(t1))

##creating tuple with a single item 
t2=(2)
print(t2); print(type(t2))

t3=('hello')
print(t3); print(type(t3))
##single item is always its own datatype


##but if we use , after that it becomes tuple

t=(4,)
print(t,type(t))
```

```python
##homogeneous tuple

t=(1,2,3,4)
print(t)

##heterogeneous tuple

x=(1,1.4,'hello',True)
print(x)


##2D tuple
t5=(1,2,3,(4,5))
print(t5)

##using type conversion

t6=tuple('hello')
print(t6)
```

Acessing items from a tuple

1-indexing

2-slicing

```python
t=(1,2,3,4)
print(t[0])
print(t[-1])
print(t[-3])

##gives index error if we give out of bound index

##slicing
print(t[0:])
print(t[::-1])
```

Operations on tuple

```python
## + and *

t1=(1,2,3,4)
t2=(5,6,7,8)

print(t1+t2)
print(t1*3)
## membership

print(1 in t1)
print(2 not in t1)
##iteration

for i in t1:
    print(i)
```

Tuple Functions

```python
#len/min/max/sorted/sum

t=(1,2,3,4)
print(len(t))
print(sum(t))
print(min(t))
print(max(t))
print(sorted(t))
print(sorted(t,reverse=True))
```

```python
#count and #index

t=(1,2,3,4,5,5)
print(t.count(5))
print(t.index(4))
print(t.index(50))
```

Difference between list and tuples

1-Syntax

list [].  tuple ()

2-mutability

tuples->immutable. lists->mutable

3-speed

tuples are faster than lists bcz they are immutable

4-Memory

tuples take lesser memory compared to lists

##immutable datatypes take lesser space than mutable datatypes

5-functions

lists have more built in functions than tuple

6-error prone

list is more error prone than tuple

7-Usability

we cannot edit in tuples (used for read only applications)



```python
import time
L=list(range(1000000))
T=tuple(range(1000000))
start=time.time()

for i in L:
    i*5
print('List time:',time.time()-start)

start=time.time()

for i in T:
    i*5
print('Tuple time:',time.time()-start)
```

```python
import sys
L=list(range(1000))
T=tuple(range(1000))
##getsizeof finds the space acquired by that in bytes
print('List size:',sys.getsizeof(L))
print('Tuple size:',sys.getsizeof(T))
```

Special syntax called as tuple unpacking 

```python
a,b,c=(1,2,3)
print(a,b,c)
```

```python
a,b=(1,2,3)
print(a,b)

##too many values to unpack
```

```python
##swapping the values

a=1
b=2
a,b=b,a ##tuple unpacking

print(a,b)
```

```python
##it is possible I only want two values from tuples and not care about other values 

a,b,*others=(1,2,3,4)
print(a,b)
print(others) ##stored in a list
```

```python
## Zipping tuples

a=(1,2,3,4)
b=(5,6,7,8)

zip(a,b) ## zip object

tuple(zip(a,b)) ##tuple of tuples
```
