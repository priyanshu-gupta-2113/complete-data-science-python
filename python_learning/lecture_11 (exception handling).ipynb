```python
class Bank:
    def __init__(self,balance):
        self.balance=balance

    def withdraw(self,amount):
        if amount<0:
            raise Exception('amount cannot be negative')
            ## Exception is class . when you raise it you create obj of Exception and that obj is send to except block 
        if self.balance < amount:
            raise Exception('gareeb')
        self.balance=self.balance-amount
    
obj=Bank(100000)

try:
    obj.withdraw(11150000)

except Exception as e:
    print(e)
else:
    print(obj.balance)
```

## Creating custom Exception

```python
class MyException(Exception):
    def __init__(self,message):
        print(message)


class Bank:
    def __init__(self,balance):
        self.balance=balance

    def withdraw(self,amount):
        if amount<0:
            raise MyException('amount cannot be negative')
            ## Exception is class . when you raise it you create obj of Exception and that obj is send to except block 
        if self.balance < amount:
            raise MyException('gareeb')
        self.balance=self.balance-amount
    
obj=Bank(100000)

try:
    obj.withdraw(-5000000)

except MyException as e:
    pass
else:
    print(obj.balance)
```

```python
## why to create our own custom exception class ?
## full control 
## application based work apart from only printing the errors 
```

```python
## example
class SecurityError(Exception):
    def __init__(self,message):
        print(message)
    def logout(self):
        print('log out')
class Google:
    def __init__(self,name,email,password,device):
        self.name=name
        self.email=email
        self.password=password
        self.device=device
    
    def login(self,email,password,device):
        if (device!=self.device):
            raise SecurityError('bhai teri toh lag gayi') ## message to user that there is wrong and logs out all devices 
        if email==self.email and password==self.password:
            print('welcome')
        else:
            print('login error')


obj=Google('nitish','nitish@gmail.com','1234','android')

try:
    obj.login('nitish@gmail.com','1234','windows')
except SecurityError as e:
    e.logout()

else:
    print(obj.name)

finally:
    print('database connection closed')

```
