## to_datetime function

converts existing objects to pandas timestamp/datetimeindex object

```python
## simple series example

s=pd.Series(['2023/1/1','2021/1/1','2025/1/1'])
s
```

```python
s.str.split('/').str.get(0)
```

```python
## convert this into datetime

pd.to_datetime(s)
```

```python
pd.to_datetime(s).dt.year
```

```python
pd.to_datetime(s).dt.month
```

```python
pd.to_datetime(s).dt.month_name()
```

```python
pd.to_datetime(s).dt.day_name()
```

```python
## with errors
s=pd.Series(['2023/1/1','2021/130/1','2025/1/1'])
s
```

```python
## pd.to_datetime(s) ## gives error

c=pd.to_datetime(s,errors='coerce') ## NaT is not a time 
c
```

```python
c.dt.year
```
