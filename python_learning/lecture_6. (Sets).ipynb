Deleting items in a set

del
discard
remove
pop
clear

```python
#del

##can delete whole set
s={1,2,3,4}
print(s)
del s
print(s)
```

```python
s={1,2,3,4}
##cannot delete particular items becasue indexing is not there
del s[0]
```

```python
##discard: we tell which item to delete

s={1,2,3,4,5}
s.discard(5)
print(s)

s.discard(6)
print(s)
```

```python
##remove:very similar to discard
s={1,2,3,4,5}
s.remove(5)
print(s)
```

```python

##remove will throw an error if we try to delete an element that doesn't exist
s={1,2,3,4,5}
s.remove(6)
```

```python
##Pop: randomly delete one item from the set

s={10,1,2,3,4,5,6}
s.pop()
print(s)
```

```python
##clear: to empty the set
s={1,2,3,4,5,6}
s.clear()
print(s)
```

set operations
1-union(|)
2-difference(-)
3-intersection(&)
4-symmetric difference(^)
5-Memebership test
6-Iteration

```python
##Union

s1={1,2,3,4,5}
s2={1,4,6,7,8}
s1 | s2
```

```python
#intersection
s1={1,2,3,4,5}
s2={1,4,6,7,8}
s1 & s2
```

```python
#difference
s1={1,2,3,4,5}
s2={1,4,6,7,8}
s1 -s2
```

```python
##symmetric difference  --> common ko chodke baki sab print hojayega
s1={1,2,3,4,5}
s2={1,4,6,7,8}
s1 ^ s2
```

```python
##membership test
s1={1,2,3,4,5}
print(1 in s1) 
1 not in s1
```

```python
##set functions - min/max/sorted/len/sum
s={1,2,3,4,4,5,6}
print(s)
print(len(s))
print(max(s))
print(min(s))
print(sum(s))
sorted(s)
```

```python
#union/update

s1={1,2,3,4,5,6}
s2= {4,5,6,7,8}

print(s1.union(s2)) ## same as s1|s2
print(s1)

s1.update(s2) ## it is union and permanent update
print(s1)
print(s2)
```

```python
##intersection and intersection_difference
s1={1,2,3,4,5}
s2={1,2,6,7}

print(s1.intersection(s2))
print(s1)

s1.intersection_update(s2)
print(s1)
```

```python
##difference and difference_update
s1={1,2,3,4,5}
s2={1,2,6,7}

print(s1.difference(s2))
print(s1)

s1.difference_update(s2)
print(s1)
```

```python
##symmetric difference/symmetric difference update
s1={1,2,3,4,5}
s2={4,5,6,7,8}

print(s1.symmetric_difference(s2))

s1.symmetric_difference_update(s2)
print(s1)
```

```python
#isdisjoint/issubset/issuperset

s1={1,2,3,4}
s2={5,6,7,8}
s3={3,4,5,6}
s4={1,2,3,4,5,6,7}

print(s1.isdisjoint(s2))
print(s1.isdisjoint(s3))
print(s4.issuperset(s1))
print(s1.issubset(s4))
```

```python
#copy -->creates shallow copy

s1={1,2,3}
s2=s1.copy()

s1.add(7)
print(s1)
print(s2)
```
