accessing items from a list

1-indexing 
2-slicing

lists work very similar to strings

```python
##indexing and slicing
L=[1,2,3,4,5]

##positive indexing left to right and starts from 0
print(L[0])
print(L[3])

##negative indexing right to left and starts from -1

print(L[-1])
print(L[-3])

##2D list

L=[1,2,3,[4,5]]

print(L[-1])
print(L[-1][-2])

print(L[3][0])

##3D list

L=[[[1,2],[3,4]],[[5,6],[7,8]]]
print(L[1])
print(L[1][0])
print(L[1][0][0])
```

```python
##Slicing 
##to extract multiple items from a list we use slicing

L=[1,2,3,4,5,6]
print(L[0:3])

print(L[0::2])

print(L[-3:])

print(L[-3::-1])

print(L[-5:-2:2])

##to reverse list 
print(L[::-1])
```

Deleting items from a list

```python
#del

L=[1,2,3,4,5]
print(L)
del L  ##deletes L from memory and NOW L has no existence
print(L)
```

```python
#del
L=[1,2,3,4,5]
##indexing
del L[-1]
print(L)

##slicing
del L[1:3]
print(L)
```

```python
##remove: Deals by value suppose you do not know where is 5 value index and only know value 5
L=[1,2,3,4,5]
L.remove(5)
##FINDS 5 AND DELETES IT 
print(L)

L=[1,2,3,4,5,5,5,5,5]
L.remove(5)
print(L)
```

```python
#pop
L=[1,2,3,4,5]

L.pop(0) ## 0 means 0th index here
print(L)

##If we don't provide index then it will delete the last elemenet
L.pop()
print(L)
```

```python
#clear function: doesn't delete the list but make the list empty
L=[1,2,3,4,5]

L.clear()
print(L)
```

```python
## Nested if with list comprehensions
basket=['apple','cherry','guava','banana']
my_fruits=['apple','kiwi','grapes','banana']

##add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

lst1=[fruit for fruit in my_fruits if fruit in basket and fruit[0]=='a']
print(lst1)


lst2=[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
print(lst2)
```

```python
##print a (3*3)matrix using list comprehension ->nested list comprehension 

[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
## cartesian product ->List comprehension or 2 list together

L1=[1,2,3,4]
L2=[5,6,7,8]

[i*j for i in L1 for j in L2]
```

Zip 

The zip() function returns a zip object, which is list of tuples where the first item in each passed list is paired together,and then the second list in each passed list are paired together.

if the passed lists have different lengths,the list with the least items decides the length of the new list

```python
##write a program to add items of 2 list index-wise
L1=[1,2,3,4]
L2=[-1,-2,-3,-4]

print(zip(L1,L2)) ##Creates zip object 
print(list(zip(L1,L2))) ##Convert it into list
[i+j for i,j in zip(L1,L2)]
```
