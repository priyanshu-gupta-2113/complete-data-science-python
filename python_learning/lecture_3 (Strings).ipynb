```python
##Slicing : To extract more than one characters
s="hello world"
## [start,end)
print(s[0:0]) ## is empty string
print(s[0:1])
print(s[0:2])
print(s[0:3])
print(s[2:3])
print(s[2:]) ## string starting from index 2
print(s[:3]) ## gives string from 0 to index 3-1=2


```

```python
##stepsize 
s='hello world'
print(s[0:5:2])
print(s[0:6:3])

##to print the string in reverse :
print(s[::-1])
print(s[6:0:-2])
print(s[6::-2])


```

```python
## String specific functions 

##1-Capitalize /Title/Upper/Lower/Swapcase

##capitalize function makes the first letter of string capital that's it ##used in database for names
s='hello world'
print(s.capitalize())

##title function sentence ke har word ka pahle character ko capital mein print kardega
print(s.title())


##Upper function makes all the characters of string uppercase and oppoisite for lower function

print(s.upper())
print('Hello WORLD'.lower())

##swapcase function makes capital letters small and small capital

print('Hello World'.swapcase())

```

```python
##write a program which can remove a particular character from string

s=input("Enter the string:")
c=input("enter the character to remove")
result=''

for i in s:
    if i !=c:
        result=result+i

print(result)
```

```python
##check if string is palindrome or not 
##abba
##malyalam

s=input('enter the range:')

for i in range(0,len(s)//2):
    if s[i]!=s[len(s)-i-1]:
        print('not palindrome')
        break
else :
    print('palindrome')
```

```python
##write a program to count the no of words in a string without split()
s=input("Enter the string:")
counter=0
for i in s:
    if i==' ':
        counter+=1

print(counter+1)
```

```python
##write a program to create the list of words in a string without split()
s=input("enter the string:")
l=[]
temp=''
for i in s:
    if i!=' ':
        temp=temp+i
    else:
        l.append(temp)
        temp=''
l.append(temp)
print(l)
```

```python
##note :temp = temp + i

## You are NOT modifying the existing string
## You are creating a NEW string and reassigning temp to it
##❌ This would violate immutability (NOT allowed):
## temp[0] = 'H'   # ❌ ERROR
```

```python
##write a python program to convert string to title case without using the title() function

s=input("enter the string:")

lst=s.split()
L=[]
for i in lst:
    ## now printing the string here 
    L.append(i[0].upper()+i[1:].lower())
print(L)
```

```python
##write a program that can convert an integer to string

number=int(input("enter the number:"))
digits='0123456789'

result=''

while number!=0:
    num=number%10
    result=digits[num]+result
    number=number//10

print(result)
```
