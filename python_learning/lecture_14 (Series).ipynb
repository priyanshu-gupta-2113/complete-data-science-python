Series from lists

```python
#string

country=['india','england','denmark','USA']

pd.Series(country) ##in pandas there is a class named series and you create object of that class series

##in series there are two things index and its value
##dtype object it means elements present in series are of datatype object(in python strings is called object)
```

```python
##integers

runs=[13,24,56,100]
pd.Series(runs)
```

```python
##custom index

marks=[56,67,86,100,80]
subjects=['maths','english','phy','chem','bio']
pd.Series(marks,index=subjects)
```

```python
##Setting a name

marks=[56,67,86,100,80]
subjects=['maths','english','phy','chem','bio']

marks=pd.Series(marks,index=subjects,name="priyanshu's marks")
marks
```

```python
##series from dict

marks={
    'maths':56,
    'english':67,
    'phy':100,
    'chem':86,
    'bio':80
}

marks_series=pd.Series(marks,name='priyanshu ke marks')
print(marks_series)
##keys become indices and values become col values
##marks_series is reference variable and priyanshu ke marks is attribute of that object
```
