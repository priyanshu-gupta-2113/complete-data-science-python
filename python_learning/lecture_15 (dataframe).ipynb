```python
points_df= pd.DataFrame({
    '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
    '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
})

points_df

## for dataframe if we apply drop then it traverse row wise instead of column wise 
```

```python
def distance(row):
    pt_a=row['1st point']
    pt_b=row['2nd point']

    return ((pt_a[0]-pt_b[0])**2 + (pt_a[1] - pt_b[1])**2)**0.5

points_df.apply(distance,axis=1)
## for implementing it across rows we always set axis =1 
```

```python
## corr
## corr() is used to find the relationship between numerical columns in a DataFrame.
## It tells how strongly two variables are related.
## synatx. df.corr()
'''
🔹 Correlation values

+1 → Perfect positive relationship

-1 → Perfect negative relationship

0 → No relationship
'''

data = {
    'Maths': [80, 70, 90, 85],
    'Science': [75, 65, 95, 80],
    'English': [78, 68, 88, 82]
}

df = pd.DataFrame(data)
print(df.corr())
```
