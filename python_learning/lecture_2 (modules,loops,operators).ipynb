```python
##Operators in Python:
##operator is a special symbol that performs specific operations on one or more operands (values or variables) and produces a result.
#Arithmetic Operators: +, -, *, /, %, **, //

print(5+6)
print(10-4)
print(3*4)
print(5/6)
print(5//6) #floor division
print(11//2)
print(2**3) #exponentiation
print(5%2)  #modulus

#Comparison Operators: ==, !=, >, <, >=, <=
print(5==6)
print(4<5)
print(7>=3)
print(2!=2)
print(3>4)
print(5<=5)

#bitwise operators: &, |, ^, ~, <<, >>.  they operate on the binary representations of integers.
print(5 & 3)  #bitwise AND. 5=101, 3=011 => 001=1
print(5 | 3)  #bitwise OR.  5=101, 3=011 => 111=7
print(5 ^ 3)  #bitwise XOR. 5=101, 3=011 => 110=6
print(~5)     #bitwise NOT. 5=0000...0101 => 1111...1010 = -6 (two's complement)
print(5 << 1) #left shift 5=101 => 1010=10
print(5 >> 1) #right shift 5=101 => 010=2



#Logical Operators: and, or, not
print(True and False)
print(True or False)
print(not True)
print(1 and 0)
print ( 1 or 0)
print(not 1)
print(not 0)
#Assignment Operators: =, +=, -=, *=, /=, %
a = 5
a += 3  # a = a + 3
print(a)
a -= 2  # a = a - 2
print(a)
a *= 4  # a = a * 4
print(a)
a /= 2  # a = a / 2
print(a)
a %= 3  # a = a % 3
print(a)

##in python, there is no ++a and a++ operators like c/c++. to increment or decrement a variable, you can use += and -= operators.
a = 5
a += 1  # equivalent to ++a or a++
print(a)
a -= 1  # equivalent to --a or a--
print(a)



#Membership Operators: in, not in.  tells whether a value is present in a sequence (like a list, tuple, or string).
my_list = [1, 2, 3, 4, 5]
print(3 in my_list)
print(6 not in my_list)
#Identity Operators: is, is not
x = [1, 2, 3]
y = x
z = [1, 2, 3]
print(x is y)      # True, because y references the same object as x
print(x is not z)  # True, because z is a different object, even though they have the same content
# --- IGNORE ---
# --- IGNORE ---
```

```python
##random module 
##Used to generate random numbers
import random
print(random.randint(1,10))
print(random.binomialvariate(5,0.5))
print(random.choice(['apple','banana','cherry']))
print(random.random())
print(random.uniform(1.5, 5.5)) #random float between 1.5 and 5.5
print(random.randrange(0, 101, 5)) #random even number between 0 and 100
print(random.shuffle([1,2,3,4,5]))  #shuffles the list in place
print(random.sample([1,2,3,4,5], 3) ) #randomly selects 3 unique elements from the list
print(random.seed(42))  #sets the seed for reproducibility
print(random.getrandbits(4))  #generates a random integer with 4 random bits
print(random.expovariate(1/5))  #exponential distribution with mean 5
print(random.gauss(0, 1))  #Gaussian distribution with mean 0 and standard deviation 1
print(random.lognormvariate(0, 0.1))  #log-normal distribution
print(random.paretovariate(2.0))  #Pareto distribution with shape parameter 2.0
print(random.weibullvariate(1.0, 1.5))  #Weibull distribution with scale 1.0 and shape 1.5
print(random.vonmisesvariate(0, 1))  #Von Mises distribution with mu=0 and kappa=1
print(random.triangular(1.0, 10.0, 5.0))  #triangular distribution with low=1.0, high=10.0, mode
```

```python

##two types of loops: for and while loop

##while loop example:program to print table
##loop example: keep accepting numbers from user until they enter 0, then print the sum of all entered numbers.
i=1
number=12
while i<=10:
    print(number,'*',i,'=',number*i)
    i+=1
```

```python
## while loop with else
i=1
number=12
while i<=10:
    print(number,'*',i,'=',number*i)
    i+=1
else:
    print("Loop completed successfully")
```

```python
##guessing game using while loop
import random
number_to_guess=random.randint(1,100)
user_guess=None
attempts=0
while(user_guess!=number_to_guess):
    user_guess=int(input("Enter your guess (1-100):"))
    attempts+=1
    if(user_guess<number_to_guess):
        print("too low! try again.")
    elif(user_guess>number_to_guess):
        print("too high! try again.")
    else:
        print(f"Congratulations! you guessed the number {number_to_guess} in {attempts} attempts.")
```

```python
##for loop
##it uses a function called range() to generate a sequence of numbers
for i in range(1,11):
    print(i)


for i in range(1,21,2):  #prints odd numbers from 1 to 20
    print(i)
for i in range(10,0,-1):  #prints numbers from 10 to 1
    print(i)

##print table of a number using for loop
number=int(input("Enter a number to print its table:"))
for i in range(1,11):
    print(number,'*',i,'=',number*i)
##for loop with else
for i in range(1,11):
    print(number,'*',i,'=',number*i)
else:
    print("Table printed successfully")
```
