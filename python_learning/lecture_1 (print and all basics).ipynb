```python
##print function arguments
'''
1-value(s): the value or values to be printed. multiple values can be separated by commas.  
2-sep: specifies the separator between multiple values. default is a space character (' ').
3-end: specifies what to print at the end of the output. default is a newline character ('\n').
4-file: specifies the file or stream to which the output should be written. default is sys.stdout (the console).
5-flush: a boolean value that specifies whether to forcibly flush the output buffer. default is False.
flush means to clear the output buffer and write any buffered data to the output stream immediately.in easy words, it ensures that the output is displayed right away rather than being delayed until the buffer is full or the program ends.       
'''
print("Hello", "World", sep=", ", end="!\n", flush=True)  # Outputs: Hello, World!      
# In this example, the separator is set to ", ", the end character is set to "!\n", and the output buffer is flushed immediately.
##print function return value
'''
The print function in Python does not return any value. It returns None.
This means that you cannot assign the result of a print statement to a variable or use it in an expression.
The primary purpose of the print function is to display output to the console or terminal, rather than to produce a value that can be used in further computations.
'''
result = print("Hello, World!")  # This will output: Hello, World!
print(result)  # This will output: None
##In summary, the print function in Python is used to display output to the console or terminal. It takes multiple arguments, including the values to be printed, a separator, an end character, a file or stream, and a flush option. The print function does not return any value and is primarily used for displaying output.
```

```python
##Dynamic Typing in Python

##in C,you cannot change the datatype of variable once it is declared. but in python you can change the datatype of variable at any time by assigning a value of different datatype to it.
x = 10          # x is an integer
print(x, type(x))  # Outputs: 10 <class 'int'>
x = "Hello"    # Now x is a string
print(x, type(x))  # Outputs: Hello <class 'str'>
x = 3.14       # Now x is a float
print(x, type(x))  # Outputs: 3.14 <class 'float'>
x = True       # Now x is a boolean
print(x, type(x))  # Outputs: True <class 'bool'>
##python interpreter is intelligent enough to handle dynamic typing and automatically manage memory allocation 
```
