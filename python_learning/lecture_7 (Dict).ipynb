Dictionary

Dictionary in python is collection of key values,used to store data values like map,which,unlike other data types which
hold only a single value as an element

In some langauges it is known as map or associative arrays

dict={'name': 'nitish', 'age':38 , 'gender': 'male'}


Characterisitics:

1-mutable

2-indexing has no meaning

3-keys cannot be duplicated

4-keys cannot be mutable items ( keys cannot be list or set )

```python
##Accessing items in a dictionary

##there is no concept of indexing
my_dict={'name':'Jack','age':26}

my_dict[0]
##[]
```

```python
my_dict={'name':'Jack','age':26}

##using keys to access
print(my_dict['name'])
print(my_dict['age'])

my_dict.get('age')
```

How to add new Key value pair

```python
my_dict={'name':'Jack','age':26}

my_dict['gender']='male'
my_dict['weight']=72

my_dict
```

Remove key value pairs

```python
d4={'name': 'Jack', 'age': 26, 'gender': 'male', 'weight': 72}
#pop -->you provide key and it removes that key-value pair

d4.pop('weight')
print(d4)


#popitem -->deletes the last item always
d4.popitem()
print(d4)


#del -->to delete whole dictionary
del d4
```

```python
#del --> can be used to delete particular key value pair

d4={'name': 'Jack', 'age': 26, 'gender': 'male', 'weight': 72}
del d4['name']

print(d4)


##clear

d4.clear()
print(d4)
```

```python
##To access 2D dictionary

s={'name':'Nitish', 'college':'bits', 'subjects':{ 'dsa':50, 'maths':40, 'eng':67}}

print(s['subjects'])
print(s['subjects']['maths'])
s['subjects']['ds']=60
s
```

```python
##Editing key-value pairs
s={'name':'Nitish', 'sem':4,'college':'bits', 'subjects':{ 'dsa':50, 'maths':40, 'eng':67}}

s['sem']=5. ##existing key and its new value
s['subjects']['dsa']=90
s
```

```python
s={'name':'Nitish', 'sem':4,'college':'bits', 'subjects':{ 'dsa':50, 'maths':40, 'eng':67}}

'Nitish' in s ##in dictionary everything is about keys not values

print('name' in s)
print('semester' not in s)
```

```python
d={'name':'nitish','gender':'male','age':33}

for i in d:
    print(i)
    ##dictionary is all about keys so it will print keys not values
```

```python
d={'name':'nitish','gender':'male','age':33}

for i in d:
    print(i,d[i])
```

Dictionary Functions


```python
##len/sorted
d={'name':'nitish','gender':'male','age':33}

len(d)

print(sorted(d)) ## sort the keys 
print(min(d)) ##gives min key
max(d)  ##gives max key
```

```python

##items/keys/values
d={'name':'nitish','gender':'male','age':33}

print(d.items()) ##displays all items in the form of list of tuples

print(d.keys()) ##gives list of keys

print(d.values())
```

```python
##update
d1={1:2,3:4,4:5}
d2={4:7,6:8}

d1.update(d2)

print(d1)
```

```python
##dictionary comprehensions

#print 1st 10 numbers and their squares

{x: x*x for x in range(1,11)}
```

```python
##using existing dictionary
distances={'delhi':1000,'mumbai':2000,'bangalore':3000}
print(distances.items())

{key:value*0.62 for (key,value) in distances.items() }
```

```python
##using zip

days=['sunday','monday','tuesday','wednesday','thursday','friday','saturday']
temp_c=[30.5,32.6,31.8,33.4,29.8,30.2,29.9]

list(zip(days,temp_c))


{i:j for (i,j) in zip(days,temp_c)}
```

```python
##using if condition

products={'phone':10,'laptop':0,'charger':32,'tablet':0}

{i:j for (i,j) in products.items() if j>0}
```

```python
##nested comprehensions
#print tables of number from 2 to 4
{

    2:{1:2,2:4,3:6,4:8,5:10},
    3:{1:3,2:6,3:9,4:12,5:15},
    4:{1:4,2:8,3:12,4:16,5:20}
}


{i: {j:i*j for j in range(1,11)}for i in range(2,5)}
```
