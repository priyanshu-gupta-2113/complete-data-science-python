```python
## split (apply) combine
## apply -> builtin function

## first we split the data by applying groupby (splits dataframe into small dataframes of a genre)
## you apply transformation to a small group by using apply function 
## then you combine the results for differnet groups


genres.apply(min) ##combines min value for all the groups 
```

```python
def foo(group):
    print(group)
    return group

genres.apply(foo)

## apply ek ek karke saare groups bhej raha hai 
## these groups are dataframes 
```

```python
## find no of movies starting with A for each group

def foo(group):
    return group['Series_Title'].str.startswith('A')

genres.apply(foo)
```

```python
## find no of movies starting with A for each group

def foo(group):
    return group['Series_Title'].str.startswith('A').sum()

genres.apply(foo)
```

```python
## find ranking of each movie in the group according to IMDB score

##khudke group mein movie ke kya rank hai ?

def ranking(group):
    group['ranking']=group['IMDB_Rating'].rank(ascending=False)
    return group

genres.apply(ranking)
```

```python
## find normalized imdb rating group wise
## through apply we can implement custom logic 
 ## x new= x- xmin / xmax- xmin

def normal(group):
    group['norm_rating']=(group['IMDB_Rating']-group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group 

genres.apply(normal)
```

## Merge

you join databases on the basis of one common column

```python
## inner join
## synatx: left_table.merge(right_table,how='type of join' ,on='common_col_name')

students.merge(registrations,how='inner',on='student_id') ## last 3 registrations are not there 
```

```python
## left join

## i want to join courses and registrations (both have course_id common)
courses.merge(registrations,how='left',on='course_id')
```

```python
## right join

temp_df=pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
}
)
temp_df
## students and registrations 

students=pd.concat([students,temp_df],ignore_index=True)
students
```

```python
students.merge(registrations,how='right',on='student_id')
```

```python
## outer join

students.merge(registrations,how='outer',on='student_id')
```
