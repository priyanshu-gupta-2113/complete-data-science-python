### `*args and **kwargs`
`*args and **kwargs` are special Python keywords that are used to pass the variable length of arguments to a function


args means arguments


```python
# *args allows us to pass a variable number of non-keyword arguments to a function.


def multiply(a,b):
    return a*b

multiply(2,3)

##kitni numbers ka input leke multiply karke dede which means you also do not know how many inputs users can give

def multiply(*args):
    product=1
    for i in args:
        product=product*i
    return product

print(multiply(1,2,3,4,4,4,4,4,4))
print(multiply(1,1,1,1,2,2,3,4))


##behind the scene:
'''
when you write * it understands there will be variable number of arguments so it internally
 creates a tuple
'''

##it stores the arguments in a tuple
```

```python
def multiply(*args):
    product=1
    for i in args:
        product=product*i
    print(args)
    return product

multiply(1,2,3,34,5,5,6,7)


##args is not a keywords u can use any keyword which you want

def add(*salmon):
    sum=0
    for i in salmon:
        sum=sum+i
    return sum

add(1,2,2,2,2,2,2,2)
```

```python
# **kwargs means keyword arguments
# **kwargs allows us to pass any number of keyword arguments.
# Keyword arguments mean that they contain a key-value pair, like a Python dictionary.


## ** will convert it into dictionary


def display(**kwargs):
    for (key,value) in kwargs.items():
        print(key,'->',value)

display(india='delhi',srilanka='colombo',nepal='katwandu')
```

##### Points to remember while using `*args and **kwargs`

- order of the arguments matter(normal -> `*args` -> `**kwargs`)
- The words args and kwargs are only a convention, you can use any name of your choice

### Lambda Function

A lambda function is a small anonymous function. (inka naam nahi hota)

A lambda function can take any number of arguments, but can only have one expression.

image.png

```python
# x -> x^2

print(lambda a:a*a)
print(type(lambda a:a*a))
x=lambda a:a*a
x(2)
```

```python
# x,y -> x+y
a=lambda x,y:x+y

a(2,3)
```

#### Diff between lambda vs Normal Function

- No name
- lambda has no return value(infact,returns a function)
- lambda is written in 1 line
- not reusable

Then why use lambda functions?<br>
**They are used with Higher order functions**


```python
# check if a string has 'a'
a=lambda s:'a' in s
a('hello')
```

```python
# odd or even
a=lambda x:'even' if x%2==0 else 'odd'
a(5)
```

### Higher Order Functions

```python
##higher order function is a function jiske return mein khud ek function hota hai
##esa function joh input mein dusre function ko receive kare

# Example

def square(x):
    return x**2

def transform(f,L):
    output=[]
    for i in L:
        output.append(f(i))

    print(output)


L=[1,2,3,4,5]
transform(square,L)
```

```python
def transform(f,L):
    output=[]
    for i in L:
        output.append(f(i))

    print(output)


L=[1,2,3,4,5]
transform(lambda x:x**2,L)

##if we want cube just do x**3

##rather than creating a cubic function just create lambda function
```

### Map

```python
# expects two things map function and iterable (list)

##example finding square

map(lambda x:x*x, [1,2,3,4,5]) ##map object

list(map(lambda x:x*x,[1,2,3,4,5]))
```

```python
# odd/even labelling of list items
list(map(lambda x:'odd' if x%2!=0 else 'even', [1,2,3,5,6,7,123,423,12,34]))
```

```python
# fetch names from a list of dict

users=[
    {    
        'name':'Rahul',
        'age':45,
        'gender':'male'
    },
    {
      'name':'Nitish',
      'age':33,
      'gender':'male'
    },
    {
     'name':'Ankita',
     'age':50,
     'gender':'female'
    }
]

list(map(lambda user:user['name'],users))
```

### Filter


given list se condition ke basis pe filtering karta hai

```python
# numbers greater than 5
## filter(lambda function which tells condition , list)
L=[3,4,5,6,7]
list(filter(lambda x:x>5,L))
```

```python
# fetch fruits starting with 'a'
fruits=['apple','guava','cherry']

list(filter(lambda x:x.startswith('a'),fruits))
```
